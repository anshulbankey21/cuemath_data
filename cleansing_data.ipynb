```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("india_cuemath_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
```

```python
df["city"] = df["city"].fillna("online")
df["tutor_name"] = df["tutor_name"].fillna("Guest_lecture")
```

```python
# Make sure the date columns are in datetime format
df['enrollment_date'] = pd.to_datetime(df['enrollment_date'])
df['last_login'] = pd.to_datetime(df['last_login'])

# Create random login offsets of 1 to 3 days
random_days = np.random.randint(1, 4, size=len(df))

# Fill missing last_login with enrollment_date + random_days
df['last_login'] = df['last_login'].fillna(df['enrollment_date'] + pd.to_timedelta(random_days, unit='D'))
```

```python
df["age"] = df["age"].fillna(df["age"].median())
```

```python
df.head()
```

```python
# Find correlations between numeric columns
correlations = df.corr(numeric_only=True)
print(correlations)
```

```python
df.drop(columns=["feedback_score.1"],inplace=True)
```

```python
df['quiz_score'] = df['quiz_score'].fillna(df['final_grade'] * 0.8)
```

```python
# Step 1: Fill weekly_progress using time_spent_minutes
df['weekly_progress(%)'] = df['weekly_progress(%)'].fillna(df['time_spent_minutes'] * 0.1)

# Step 2: Fill time_spent_minutes using weekly_progress (after some are filled in step 1)
df['time_spent_minutes'] = df['time_spent_minutes'].fillna(df['weekly_progress(%)'] * 10)

# Step 3: Fill feedback_score using parent_rating
df['feedback_score'] = df['feedback_score'].fillna(df['parent_rating'])

# Step 4: Fill parent_rating using feedback_score (after some are filled in step 3)
df['parent_rating'] = df['parent_rating'].fillna(df['feedback_score'])

# Step 5: Still remaining nulls? Fill with a default or average
df['weekly_progress(%)'] = df['weekly_progress(%)'].fillna(df['weekly_progress(%)'].mean())
df['time_spent_minutes'] = df['time_spent_minutes'].fillna(df['time_spent_minutes'].mean())
df['feedback_score'] = df['feedback_score'].fillna(3)  # Neutral feedback
df['parent_rating'] = df['parent_rating'].fillna(3)    # Neutral rating
```
